```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# bar plot
tips = sns.load_dataset('tips')
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("average total bill by day")
plt.show()
```

```python
# count plot
sns.countplot(x='sex', data=tips)
sns.set_style('darkgrid')
plt.title("gender count")
plt.show()
```

```python
# heatmap
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')
plt.title("passengers by month and year")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

data = pd.read_csv('/content/drive/MyDrive/Rahi_AIML/2.Data science/4_seaborn/tested.csv')

data = pd.read_csv("/content/drive/MyDrive/Rahi_AIML/2.Data science/4_seaborn/tested.csv")

print(data.head())

sns.countplot(x='Survived', data=data)
plt.title("survival count")
plt.xlabel("survived (0=NO, 1=YES)")
plt.ylabel("number of passengers")
plt.show()
```

```python
sns.countplot(x='Sex', data=data, palette='Set2')
plt.title("gender distribution")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=data, palette='husl')
plt.title("Survival by Gender")
plt.show()
```

```python
plt.hist(data['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=data, palette='coolwarm')
plt.title("Fare by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
# Select only numeric columns before calculating the correlation
numeric_data = data.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title("feature correlation")
plt.show()
```

```python
sns.pairplot(data[['Survived', 'Pclass', 'Age', 'Fare']].dropna(), hue='Survived', palette='Set1')
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```
