common plot types in matplotlib:
line plot -- plt.plot()
bar chart -- plt.bar()
histogram -- plt.hist()
scatter plot -- plt.scatter()
pie chart -- plt.pie()

```python
import matplotlib.pyplot as plt
# line plot
x=[1,2,3,4,5]
y=[10,12,8,14,7]
plt.plot(x,y)
plt.title("line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# bar plot
subjects=["math","science","english"]
marks=[90,50,75]
plt.bar(subjects,marks, color='skyblue')
plt.title("student marks")
plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

```python
# scatter plot
height = [150, 160, 170, 180]
weight = [50, 60, 65, 80]

plt.scatter(height, weight, color='red')
plt.title("Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

 Customization Options
Color: color='red'

Line Style: linestyle='--', ':', '-.'

Markers: marker='o', '*', 's'

Grid: plt.grid(True)

Legend: plt.legend()

```python
plt.plot(x,y,color='green', marker='*', linestyle='--', label='data line')
plt.legend()
plt.grid(True)
plt.show()
```
